#Read processed dataset-
All_Sentences.csv, generated in BaselineDataPrep

```python

df_copy = pd.read_csv('/content/drive/MyDrive/ENGI 4800/data/All_Sentences.csv', index_col=False).dropna()[['Original_txt','AuthorsId_removed', 'punctuation_stopWord_removed','stemmed']]
m,n = df_copy.shape
df_copy.index = np.arange(m)
df_copy.columns = ['txt','AuthorsId_removed', 'punctuation_stopWord_removed','stemmed']

df_copy.info()
df = df_copy.copy().drop(['AuthorsId_removed', 'punctuation_stopWord_removed', 'stemmed'], axis=1)
```

##Term Search -- Investigation

Investigation	Conducting a research and investigation process, specifically performing the experiments, or data/evidence collection

```python
#filter standard
df_investigation = df.loc[df.txt.str.contains('investigat',na=False) | 
                              df.txt.str.contains('research',na=False) | 
                              df.txt.str.contains('experiment',na=False) |
                              df.txt.str.contains('collect',na=False)]
```

```python
df_investigation.info()
df_investigation.head(5)
```

```python
df_investigation = df_investigation.sample(n=100)
df_investigation.to_csv('/content/drive/MyDrive/ENGI 4800/data/Investigation_Matched_Sentences.csv')
```
